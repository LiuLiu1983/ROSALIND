```python
"""count each seq, add the information to profile matrix"""
def count_seq(seq, matrix):
    seq= seq.upper()
    if len(seq) != len(matrix[0]):
        print(len(seq))
        #print(matrix[0])
        #raise ValueError('length not right')
    for i in range(len(seq)):
        if seq[i]=="A":
            matrix[0][i]+=1
        elif seq[i]=="C":
            matrix[1][i]+=1
        elif seq[i]=="G":
            matrix[2][i]+=1           
        elif seq[i]=="T":
            matrix[3][i]+=1
    return matrix

"""read fasta file"""
filename="\\rosalind_cons.txt"
dictionary="D:"
with open(dictionary+filename,"r") as f:
    line=f.readline().strip()
    seqline=''
    if line[0]==">": 
        line=f.readline().strip()
    while line[0]!=">":
        seqline+=line
        line=f.readline().strip()
    length= len(seqline)
    profile_matrix = [[0]*length for _ in range(4)]
    profile_matrix = count_seq(seqline, profile_matrix)
    seqline=f.readline().strip() 
    for line in f:
        line=line.strip()
        if line[0]==">":
            profile_matrix = count_seq(seqline, profile_matrix)
            seqline=''
        else:
            seqline+=line
    profile_matrix = count_seq(seqline, profile_matrix)
            
            
matrix= profile_matrix
print(matrix)
```

```python
"""go through the profile matrix and find the largest number for each position and append the corresponce character to the consensus string"""
consensus_string=''
for i in range(len(matrix[0])):
        max_pos=-1
        index=-1
        for j in range(len(matrix)):
            if matrix[j][i]>=max_pos:
                max_pos = matrix[j][i]
                index=j
    
        if index == 0:
            consensus_string+='A'
        elif index == 1:
            consensus_string+='C'    
        elif index == 2:
            consensus_string+='G'       
        elif index == 3:
            consensus_string+='T'           

print(consensus_string)
#print(matrix)
for i in range(len(matrix)):
    num_str = " ".join(map(str,matrix[i]))
    if i == 0:
        print("A: "+num_str)
    elif i ==1:
        print("C: "+num_str)        
    elif i ==2:
        print("G: "+num_str)        
    elif i ==3:
        print("T: "+num_str)    
```
