```python
def two_way_partition(arr):
    arr[0],arr[-1]= arr[-1],arr[0]
    pivot = arr[-1]
    i = 0
    j = -1
    
    for i in range(0,len(arr)):
        if arr[i]<=pivot:
            j+=1
            arr[i], arr[j] = arr[j], arr[i]
    return arr
        
            
```
