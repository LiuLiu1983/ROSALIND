```python
def call_ORF(seq):
    """return a list of sub_seq that be called as ORF, 
    from both the sense starnd and the reverse complement."""
    start_code= ["ATG"]
    end_code = ['TAA', 'TAG', 'TGA']
    ORF_list = []
    for i in range(len(seq)):
        if i+3 <= len(seq) and seq[i:i+3]==start_code[0]:
            #print(i)
            for j in range(i+3,len(seq),3):
                if seq[j:j+3] in end_code:
                    ORF_call = seq[i:j+3]
                    ORF_list.append(ORF_call)
                    break
    #print(ORF_list)                
    """remove the replicate seq in ORF_list"""
#     redundant_list = []
#     for seq in ORF_list:
#         for compare in ORF_list:
#             if seq in compare and seq != compare:
#                 redundant_list.append(seq)
#     ORF_list = list(set(ORF_list)-set(redundant_list))
    return ORF_list
        

def complement_DNA(seq):
    """return the reverse complement DNA"""
    complement_seq = ""
    complement_code={"A":"T","T":"A","C":"G","G":"C"}
    for i in range(len(seq)-1, -1, -1):
        complement_seq+=complement_code[seq[i]]
    return complement_seq


def call_ORF_frombothstrand(seq):
    ORF_list = []
    complement_seq = complement_DNA(seq)
    ORF_list= call_ORF(seq) + call_ORF(complement_seq)
    return ORF_list
    
```

```python
codons = {'F': ['UUU', 'UUC'],
  'L': ['UUA', 'UUG', 'CUU', 'CUC', 'CUA', 'CUG'],
  'S': ['UCU', 'UCC', 'UCA', 'UCG', 'AGU', 'AGC'],
  'Y': ['UAU', 'UAC'],
  '*': ['UAA', 'UAG', 'UGA'],
  'C': ['UGU', 'UGC'],
  'W': ['UGG'],
  'P': ['CCU', 'CCC', 'CCA', 'CCG'],
  'H': ['CAU', 'CAC'],
  'Q': ['CAA', 'CAG'],
  'R': ['CGU', 'CGC', 'CGA', 'CGG', 'AGA', 'AGG'],
  'V': ['GUU', 'GUC', 'GUA', 'GUG'],
  'A': ['GCU', 'GCC', 'GCA', 'GCG'],
  'D': ['GAU', 'GAC'],
  'E': ['GAA', 'GAG'],
  'G': ['GGU', 'GGC', 'GGA', 'GGG'],
  'I': ['AUU', 'AUC', 'AUA'],
  'M': ['AUG'],
  'T': ['ACU', 'ACC', 'ACA', 'ACG'],
  'N': ['AAU', 'AAC'],
  'K': ['AAA', 'AAG']}

def change_U_to_A(seq):
    return seq.replace("U","T")

translation_code={}
for key, item in codons.items():
    for code in item:
        translation_code[change_U_to_A(code)]=key
        
translation_code
```

```python
def translate(seq_list):
    translation_list=[]
    for seq in seq_list:
        protein_seq = ""
        for i in range(0,len(seq),3):
            protein_seq += translation_code[seq[i:i+3]]
        
        if protein_seq[-1]=="*":
            translation_list.append(protein_seq[:-1])
        else:
            translation_list.append(protein_seq)
            
            
    return list(set(translation_list))
        
```
