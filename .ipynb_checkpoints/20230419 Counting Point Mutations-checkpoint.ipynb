```python
def hamming_dis(s,t):
    if len(s)!=len(t):
        raise ValueError("length is s and t is not equal")
    dis_num=0
    for i in range(len(s)):
        if s[i]!=t[i]:
            dis_num+=1
        
    return dis_num
        
```
