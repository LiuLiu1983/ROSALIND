```python
def merge_two_sorted_arrays(l1,length1,l2,length2):
    count1, count2 = 0, 0
    merged_list = []
    while count1<=length1-1 and count2<=length2-1:
        if l1[count1]<=l2[count2]:
            merged_list.append(l1[count1])
            count1 += 1
        else:
            merged_list.append(l2[count2])
            count2 += 1
    
    if count1==length1:
        merged_list+=l2[count2:]
    elif count2 == length2:
        merged_list+=l1[count1:]
    return merged_list
        
        
```

```python
with open("/Users/newuser/Downloads/rosalind_mer.txt","r") as f:
    length1 = int(f.readline().strip())
    l1 = list(map(int,f.readline().strip().split()))
    length2 = int(f.readline().strip())
    l2 = list(map(int,f.readline().strip().split()))
    results = merge_two_sorted_arrays(l1,length1,l2,length2)
    for num in results:
        print(num, end=" ")
```
