```python
def insert_sort(length, sample_list):
    swaps=0
    for i in range(1,length):
        k = i
        while k>=1 and sample_list[k-1]>sample_list[k]:
            sample_list[k],sample_list[k-1]=sample_list[k-1],sample_list[k]
            k-=1
            swaps+=1
            #print(sample_list)
    return swaps,sample_list

        
```
