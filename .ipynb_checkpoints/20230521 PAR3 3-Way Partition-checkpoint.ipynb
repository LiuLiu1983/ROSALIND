```python
def three_way_partition(arr):
    arr[0],arr[-1]= arr[-1],arr[0]
    pivot = arr[-1]
    if pivot == arr[0]:
        pivot_point = 0
    else:
        pivot_point = -1
    j = -1
    
    for i in range(0,len(arr)):
        if arr[i]==pivot:
            j+=1
            arr[i], arr[j] = arr[j], arr[i]
        elif arr[i]<pivot:
            if pivot ==-1:
                j+=1
                arr[i], arr[j] = arr[j], arr[i]                
            else:
                arr[i], arr[pivot_point] = arr[pivot_point], arr[i]
                j+=1
                arr[i], arr[j] = arr[j], arr[i]
                pivot_point+=1
        #print(i,j,pi)
        #print(arr)
    return arr
```
