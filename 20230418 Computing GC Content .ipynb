```python
def CGcalculator(string):
    CG_number = 0
    string = string.upper()
    print(string)
    for i in range(len(string)):
        if string[i] not in ["A", "T", "C", "G"]:
            raise CustomException("invalid character, not DNA sequence")
        if string[i] in ["C", "G"]:
            CG_number += 1
    print(CG_number)
    print(len(string))
    return 100*float(CG_number)/float(len(string))

directory_path = "D:"
file_path = directory_path + "\\rosalind_gc.txt"

with open(file_path, "r") as file:
    # Initialize variables
    biggest_cg = 0
    biggest_cg_name = ""
    tem_cg = 0
    tem_name=''
    tem_seq = 'A'
    readseq = False
    # Loop over each line in the file
    for line in file:
        line = line.strip()   
        if line[0]=='>':
            tem_cg = CGcalculator(tem_seq)
            tem_seq = ''
            if tem_cg > biggest_cg:
                biggest_cg= tem_cg
                biggest_cg_name = tem_name
            tem_name = line[1:]
        else:
            tem_seq += line
    
    tem_cg = CGcalculator(tem_seq)
    if tem_cg > biggest_cg:
        biggest_cg= tem_cg
        biggest_cg_name = tem_name    
            
print(biggest_cg_name)     
print(biggest_cg)
```
